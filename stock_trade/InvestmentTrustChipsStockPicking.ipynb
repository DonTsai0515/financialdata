```python
import requests 
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import datetime
import random
import time
```

```python
week_url = 'https://tw.stock.yahoo.com/rank/investment-trust-buy?period=week'
res = requests.get(week_url)
week_html = res.text
soup = BeautifulSoup(week_html, 'lxml')
html_str = soup.prettify()
```

```python
last_week_name_tag = soup.select('div[class="Lh(20px) Fw(600) Fz(16px) Ell"]')
last_week_name_list = []
for i in last_week_name_tag:
    last_week_name_list.append(i.string)
print(last_week_name_list)
```

```python
IDS = soup.find_all('div',class_='D(f) Ai(c)')
ID_list = []

for ID in IDS[1:]:
    ID=ID.find('span',class_="Fz(14px) C(#979ba7) Ell").get_text()
    ID_list.append(ID)

ID_list = [ID_list[i][0:5] for i in range(len(ID_list))]

for i in range(len(ID_list)):
    ID_list[i]= ID_list[i].replace('.','')
print(ID_list)
```

```python
last_week_numbers_of_deal_tag = soup.select('div[class="Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(78px)"] span[class="Jc(fe)"]')

last_week_numbers_of_deal_list = []
for i in range(len(last_week_numbers_of_deal_tag)):
    if 'M' in last_week_numbers_of_deal_tag[i].string:
        x = float(last_week_numbers_of_deal_tag[i].string.replace('M',''))
        x = x*(10**6)
    else:
        x = int(last_week_numbers_of_deal_tag[i].string.replace(',',''))
    last_week_numbers_of_deal_list.append(x)

last_week_numbers_of_deal_list = [last_week_numbers_of_deal_list[i:i+4] for i in range(0,len(last_week_numbers_of_deal_list),4)]
for i in range(0,(len(last_week_numbers_of_deal_list))):
    last_week_numbers_of_deal_list[i][1] = last_week_numbers_of_deal_list[i][2] #投信買超
    last_week_numbers_of_deal_list[i][2] = round(last_week_numbers_of_deal_list[i][2]/last_week_numbers_of_deal_list[i][0],3) #上週投信籌碼集中度

columns=['上週投信買張數','上週投信買賣超','上週籌碼集中度','上週成交量']
last_week_performance = pd.DataFrame(last_week_numbers_of_deal_list,columns=columns
                                    ,index = last_week_name_list)
last_week_performance['股票代碼'] = ID_list
last_week_performance = last_week_performance.head(80)
```

```python
#------------------------股票當日狀況----------------------
```

```python
day_url = 'https://tw.stock.yahoo.com/rank/investment-trust-buy?period=day'
res = requests.get(day_url)
day_html = res.text
soup = BeautifulSoup(day_html, 'lxml')
html_str = soup.prettify()
```

```python
name_tag = soup.select('div[class="Lh(20px) Fw(600) Fz(16px) Ell"]')
name_list = []
for i in name_tag:
    name_list.append(i.string)
print(name_list)
```

```python
# 'div[class="Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(78px)"] span[class="Jc(fe)"]'
numbers_of_deal_tag = soup.select('div[class="Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(78px)"] span[class="Jc(fe)"]')
numbers_of_deal_list = []

for i in range(len(numbers_of_deal_tag)):
    x = int(numbers_of_deal_tag[i].string.replace(',',''))
    numbers_of_deal_list.append(x)

numbers_of_deal_list = [numbers_of_deal_list[i:i+4] for i in range(0,len(numbers_of_deal_list),4)]
for i in range(0,(len(numbers_of_deal_list))):
    numbers_of_deal_list[i][1] = numbers_of_deal_list[i][2] #投信買超
    numbers_of_deal_list[i][2] = round(numbers_of_deal_list[i][2]/numbers_of_deal_list[i][0],3) #當日投信籌碼集中度

day_performance = pd.DataFrame(numbers_of_deal_list,columns=['投信買張數','投信買賣超','籌碼集中度','成交量']
                                    ,index = name_list)
chips = pd.merge(day_performance, last_week_performance, left_index=True, right_index=True)
chips = chips[['股票代碼','成交量','上週成交量','投信買賣超','投信買張數','籌碼集中度','上週投信買賣超','上週投信買張數','上週籌碼集中度']]
chips['投信積極度'] = round(chips['投信買賣超']/chips['上週投信買賣超'],3)
chips = chips.head(50)
chips = chips[chips['成交量']>3000]
chips.head()
```

```python
chips = chips.drop(['上週成交量','籌碼集中度'],axis=1)
chips = chips.sort_values(axis=0,by=['投信積極度'],ascending=False)
```
